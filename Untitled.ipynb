```python
import pandas as pd
import numpy as np
```

```python
file1 = 'citibikedatajan2017.csv'
file2 = 'citibikedatajan2018.csv'
file3 = '201801-citibike-tripdata.csv'
file4 = '201701-citibike-tripdata.csv'
```

```python
df1 = pd.read_csv(file1, encoding="ISO-8859-1")
df2 = pd.read_csv(file2, encoding="ISO-8859-1")
df3 = pd.read_csv(file3, encoding="ISO-8859-1")
df4 = pd.read_csv(file4, encoding="ISO-8859-1")
```

```python
df4.head()
```

```python
df3 = df3.rename(columns={"starttime":"Date"})
```

```python
df3.head()
```

```python
df4 = df4.rename(columns={"Start Time":"Date"})
```

```python
df4.head()
```

```python
year17 = pd.merge(df1, df4, on = "Date")
```

```python
year18 = pd.merge(df2, df3, on = "Date")
```

```python
year17.drop(columns=["Stop Time"])
```

```python
year17["Gender"] = year17["Gender"].astype(str)
```

```python
year17["Gender"] = year17["Gender"].replace([str(0),str(1),str(2)],["Unknown","Male","Female"])
```

```python
year18.drop(columns=["stoptime"])
```

```python
year18["gender"] = year18["gender"].astype(str)
```
